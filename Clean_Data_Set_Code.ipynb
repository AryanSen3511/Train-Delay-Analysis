```python
df['Average_Delay_Minutes'] = df['Average_Delay_Minutes'].fillna(df['Average_Delay_Minutes'].median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_delayed = df.groupby('Train_Number')['Average_Delay_Minutes'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_delayed.index, y=top_delayed.values)
plt.xticks(rotation=45)
plt.title("Top 10 Most Delayed Trains")
plt.show()
```
